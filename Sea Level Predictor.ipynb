```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python

def draw_plot():
    
    # Read data from file
    sea_level_data = pd.read_csv('epa-sea-level.csv')


    # scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(sea_level_data['Year'], sea_level_data['CSIRO Adjusted Sea Level'], label="CSIRO Adjusted Sea Level")


    # first line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(sea_level_data['Year'], sea_level_data['CSIRO Adjusted Sea Level'])
    years_extended = pd.Series(range(1880, 2051))  # Extend the years up to 2050
    plt.plot(years_extended, intercept + slope * years_extended, label='Best fit line (1880-2050)', color='red')

    # second line of best fit
    recent_data = sea_level_data[sea_level_data['Year'] >= 2000]
    slope_recent, intercept_recent, _, _, _ = linregress(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'])
    years_recent = pd.Series(range(2000, 2051))
    plt.plot(years_recent, intercept_recent + slope_recent * years_recent,label='Best fit line (2000-2050)', color='green')
    

    # labels and title
    plt.xlabel('Year')
    plt.ylabel('CSIRO Adjusted Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.legend()
```

```python
print(draw_plot())
```
